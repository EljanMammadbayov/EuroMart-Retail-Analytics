```python
import pandas as pd
```

```python
df_products = pd.read_csv(r'C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/raw/products.csv')

df_products
```

```python
#1 Standardize stock_status
df_products_clean = df_products.copy()

# Map all variations to standard values
stock_status_mapping = {
    'in stock': 'In Stock',
    'IN STOCK': 'In Stock',
    'In stock': 'In Stock',
    'low stock': 'Low Stock',
    'LOW STOCK': 'Low Stock',
    'Low stock': 'Low Stock',
    'out of stock': 'Out of Stock',
    'OUT OF STOCK': 'Out of Stock'
}

df_products_clean['stock_status'] = df_products_clean['stock_status'].replace(stock_status_mapping)

#2 Handle missing cost_price
# Option 1: Leave as NULL and document
# Option 2: Estimate from unit_price (e.g., 70% of unit_price)
df_products_clean['cost_price_imputed'] = df_products_clean['cost_price'].isna()
df_products_clean['cost_price'] = df_products_clean['cost_price'].fillna(
    df_products_clean['unit_price'] * 0.70
)

# Standardize category and subcategory (proper case). Just in case, should be no problems there.
df_products_clean['category'] = df_products_clean['category'].str.title()
df_products_clean['subcategory'] = df_products_clean['subcategory'].str.title()

print("✓ Products table cleaned!")
```
