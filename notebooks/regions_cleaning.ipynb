```python
df_regions_clean = df_regions.copy()

#1 Standardize country (proper case)
df_regions_clean['country'] = df_regions_clean['country'].str.title()

#2 Standardize region_name
region_name_mapping = {
    'benelux': 'Benelux',
    'BENELUX': 'Benelux',
    'Benelux': 'Benelux',
    'western europe': 'Western Europe',
    'Western EU': 'Western Europe',
    'Western Europe': 'Western Europe',
    'central europe': 'Central Europe',
    'Central Europe': 'Central Europe'
}
df_regions_clean['region_name'] = df_regions_clean['region_name'].replace(region_name_mapping)

#3 Standardize manager names (title case)
df_regions_clean['manager'] = df_regions_clean['manager'].str.title()

# ============================================
#4 HANDLE DUPLICATE REGIONS
# ============================================

def create_region_merge_mapping(df):
    """Create mapping for duplicate regions"""
    mapping = {}
    
    for (country, region, mgr), group in df.groupby(['country', 'region_name', 'manager']):
        if len(group) > 1:
            sorted_group = group.sort_values('region_id')
            primary = sorted_group.iloc[0]['region_id']
            duplicates = sorted_group.iloc[1:]['region_id'].tolist()
            
            for dup in duplicates:
                mapping[dup] = primary
            
            print(f"{country} - {region}: Keep {primary}, merge {duplicates}")
    
    return mapping

    # Identify duplicates and create mapping
region_id_mapping = create_region_merge_mapping(df_regions_clean)

# Remove duplicate regions
regions_to_remove = list(region_id_mapping.keys())
df_regions_clean = df_regions_clean[~df_regions_clean['region_id'].isin(regions_to_remove)]

# ============================================
# SAVE OUTPUTS FOR NEXT NOTEBOOK
# ============================================

# Save cleaned regions
df_regions_clean.to_csv('C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/regions_cleaned.csv', index=False)
print("✓ Saved: C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/regions_cleaned.csv")

# Save region mapping for orders notebook
pd.DataFrame(list(region_id_mapping.items()), 
             columns=['old_region_id', 'new_region_id']).to_csv(
    'C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/region_id_mapping.csv', index=False
)
print("✓ Saved: C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/region_id_mapping.csv")

# Also save as Python variable for easy import
import pickle
with open('C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/region_id_mapping.pkl', 'wb') as f:
    pickle.dump(region_id_mapping, f)
print("✓ Saved: C:/Users/Eljan Mammadbayov/OneDrive/Рабочий стол/Python Practice/Euromart Claude/region_id_mapping.pkl")

print("✓ Regions table cleaned!")
```
